```python
likert_disagree = {
    1: "Strongly Disagree",
    2: "Disagree",
    3: "Neutral",
    4: "Agree",
    5: "Strongly Agree"
}
likert_useful = {
    1: "Not useful",
    2: "Slightly useful",
    3: "Moderately useful",
    4: "Very useful",
    5: "Extremely useful"
}
likert_confident = {
    1: "Not confident",
    2: "Slightly confident",
    3: "Moderately confident",
    4: "Very confident",
    5: "Extremely confident"
}
```

```python
def likert_bar_chart(col):
    title = questions[col]
    answerkey = title.split("?")[-1].lower()

    # Pick the right scale dict
    if "agree" in answerkey:
        likert = likert_disagree
    elif "useful" in answerkey:
        likert = likert_useful
    elif "confident" in answerkey:
        likert = likert_confident
    else:
        print(f"Unmatched Likert type for: {title}")
        likert = likert_disagree  # sensible fallback

    # Ensure numeric, count, and enforce 1–5 ordering
    counts = (
        df[col]
        .astype(int)
        .value_counts()
        .reindex([1, 2, 3, 4, 5], fill_value=0)  # fixed order
        .rename_axis("Answer")
        .reset_index(name="Count")
    )

    counts["Label"] = counts["Answer"].map(likert)

    fig = px.bar(
        counts,
        x="Answer",          # numeric scale 1–5
        y="Count",
        text="Count",
        title=title,
    )

    fig.update_xaxes(
        range=[0.5, 5.5],
        tickvals=[1, 2, 3, 4, 5],
        ticktext=[likert[i] for i in [1, 2, 3, 4, 5]],
        title_text=""        # Likert labels are enough
    )

    fig.update_yaxes(title_text="Number of respondents")

    fig.update_traces(
        textposition="outside",
        marker_line_width=0
    )

    fig.update_layout(
        uniformtext_minsize=10,
        uniformtext_mode="hide",
        bargap=0.2
    )

    return fig
```

```python
for col in ['Q6','Q7','Q8','Q9','Q10','Q19','Q22','Q28','Q29','Q30']:
    fig = likert_bar_chart(col)
    fig.show()
```

```python
def likert_summary(series):
    s = series.astype(int)

    pct_low = (s <= 2).mean() * 100
    pct_high = (s >= 4).mean() * 100
    mean_val = s.mean()

    return pct_low, pct_high, mean_val
```
